```python
stations = pd.read_csv('station_information.csv')
stations.head()
```

```python
bike_stations = set(bikeshare.from_station_name).union(set(bikeshare.to_station_name)) 
info_stations = set(stations.name)
diff = bike_stations - info_stations - set([np.nan]) ## appears in bike_station but not in station information
print(diff)
```

```python
!pip install fuzzywuzzy
from fuzzywuzzy import fuzz
```

```python
similarity = {}
for station in diff:
    ratio = {}
    for s in info_stations:
        ratio[s] = fuzz.ratio(s, station)
    similarity[(station, max(ratio, key = ratio.get))] = ratio[max(ratio, key = ratio.get)]

#similarity
final_station_name = {}
for comb in similarity:
    if similarity[comb] >= 90: ##90% confident that two locations are the same
        final_station_name[comb[0]] = comb[1]
final_station_name
```

```python
bikeshare.from_station_name = bikeshare.from_station_name.apply(lambda x: final_station_name[x] if x in final_station_name else x)
bikeshare.to_station_name = bikeshare.to_station_name.apply(lambda x: final_station_name[x] if x in final_station_name else x)
new_df = bikeshare[bikeshare.from_station_name.apply(lambda x: x in info_stations) & bikeshare.to_station_name.apply(lambda x: x in info_stations)].reset_index(drop = True)
new_df.head(30)
```

```python
#Calculate distance (displacement), use Manhatten distence and haversine formula
import math
def haversine(lon1, lat1, lon2, lat2):
    lon1, lat1, lon2, lat2 = map(math.radians, [lon1, lat1, lon2, lat2])
    delta_lon = lon2 - lon1
    delta_lat = lat2 - lat1
    a = math.sin(delta_lat / 2) ** 2 + math.cos(lat1) * math.cos(lat2) * math.sin(delta_lon / 2) ** 2
    result = 2 * 6378 * math.asin(math.sqrt(a))
    return result

#haversine(stations.lon[0],stations.lon[1],stations.lat[0],stations.lat[1])
distance = []
station_used = {}
for i in range(len(stations)):
    station_used[stations.name[i]] = (stations.lon[i], stations.lat[i])

angel = math.radians(20)
c = math.sin(angel) + math.cos(angel)
for i in range(len(new_df)):
    straight = haversine(station_used[new_df.from_station_name[i]][0], station_used[new_df.from_station_name[i]][1], station_used[new_df.to_station_name[i]][0], station_used[new_df.to_station_name[i]][1])
    distance.append(straight * c)

new_df['distance(km)'] = distance

new_df.head(30)
```

```python
from bs4 import BeautifulSoup
import urllib
import urllib.request
import re
```

```python
def getWeather(year, month):
    data = []
    web = urllib.request.urlopen('https://climate.weather.gc.ca/climate_data/daily_data_e.html?StationID=51459&timeframe=2&StartYear=1840&EndYear=2019&Day=22&Year=' + str(year) + '&Month=' + str(month))
    soup = BeautifulSoup(web, 'html.parser')
    table = soup.find('table', attrs={'class': 'table table-striped table-hover align-cells-right data-table'})
    results = table.find('thead').find('tr')
    titles = results.find_all('th')
    title_list = []
    for title in titles[1:]:
        title = title.find('a')
        title = title.get_text().strip()
        if title != None:
            title_list.append(title)
    #title_list
    table_data = table.find('tbody')
    row_data = table_data.find_all('tr')
    row_list = []
    for row in row_data[:-4]:
        data = row.find_all('td')
        row_list.append([data[i].get_text().strip() for i in range(len(data))])
    #row_list
    temp_data = pd.DataFrame(row_list, columns = title_list)
    temp_data['date'] = [date for date in range(1, len(temp_data) + 1)]
    temp_data.date = temp_data.date.apply(lambda x: datetime.datetime(year = year, month = month, day = x))
    return temp_data

#weather2016 = [getWeather(2016, i) for i in range(7,13)]
#weather2016 = pd.DataFrame(weather2016) 
```

```python
new_bikeshare['trip_start_time'] = pd.to_datetime(new_bikeshare['trip_start_time'])
new_bikeshare
```

```python
hourly_casual = new_bikeshare[new_bikeshare['user_type']=='Casual'].groupby(pd.Grouper(key='trip_start_time',freq='h'))['user_type'].count()
hourly_casual
```

```python
hourly_member = new_bikeshare[new_bikeshare['user_type']=='Member'].groupby(pd.Grouper(key='trip_start_time',freq='h'))['user_type'].count()
```

```python
hourly_member
```

```python
hourly = pd.DataFrame(pd.date_range(start='1/10/2016',end='12/31/2016',freq='h',closed='left'))
hourly = pd.merge(hourly,pd.DataFrame(hourly_casual), how ='outer',left_on=0,right_on='trip_start_time' )
hourly = pd.merge(hourly,pd.DataFrame(hourly_member), how ='outer',left_on=0,right_on='trip_start_time' )
hourly.columns=['trip_start_time','casuals','members']
hourly.set_index('trip_start_time',inplace=True)
hourly.fillna(0,inplace=True)
```

```python
hourly.head()
```

```python
import plotly
from plotly import tools
from plotly import graph_objs as go
import plotly.io as pio
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=hourly.index,y=hourly.members,name='hourly_members'))
fig.add_trace(go.Scatter(x=hourly.index,y=hourly.casuals,name='hourly_casuals'))
fig.show()
```

```python
daily = hourly.resample('D').sum()
weekly = hourly.resample('W').sum()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=weekly.index,y=weekly.members,name='hourly_members'))
fig.add_trace(go.Scatter(x=weekly.index,y=weekly.casuals,name='hourly_casuals'))
fig.show()
```

```python
hourly_pivot = hourly.pivot_table(['casuals','members'],index=hourly.index.date,columns=hourly.index.hour,fill_value=0)
hourly_pivot = hourly.pivot_table(index=hourly.index.date,columns=hourly.index.hour,fill_value=0)
hourly_pivot.head()
```

```python
from sklearn.decomposition import PCA
```

```python
X_hours = hourly_pivot.values
X_hours
```

```python
pca_val = PCA('mle').fit_transform(X_hours)
pca_val
```

```python
from sklearn.mixture import GaussianMixture as GMM
gmm = GMM(3,'full')
gmm.fit(pca_val)
cluster_label = gmm.predict(pca_val)
```

```python
hourly_pivot['cluster'] = cluster_label
hourly = hourly.join(hourly_pivot['cluster'], on = hourly.index.date)
hourly.head()
```

```python
by_hour = hourly.groupby(['cluster', hourly.index.time]).mean()
by_hour
```

```python
dayofweek = pd.to_datetime(hourly_pivot.index).dayofweek
fig = go.Figure()
fig.add_trace(go.Scatter3d(x = pca_val[:,0], y=pca_val[:,1], z=pca_val[:,2], text = hourly_pivot.index.astype(str)+' Cluster: '+cluster_label.astype(str), hovertemplate="<b>%{text}</b><br><br>", name='', mode='markers', marker=dict(size=3, color=dayofweek, colorscale=[[0, "rgb(31,120,180)"], [1/6,"rgb(166,206,227)"],[2/6,"rgb(178,223,138)"], [3/6,"rgb(51,160,44)"], [4/6,"rgb(251,154,153)"], [5/6, "rgb(227,26,28)"],
[6/6, "rgb(240,0,0)"]],                           
colorbar=dict(tick0=0,tickmode='array',
tickvals=[0,1,2,3,4,5,6],
ticktext=['MON','TUE','WED','THU','FRI','SAT','SUN'],
thickness=20))))
fig.show()
```
